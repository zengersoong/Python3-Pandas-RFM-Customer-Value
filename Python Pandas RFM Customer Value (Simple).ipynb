```python
# Using simple lambda functions to create a new header
# E.G breaking up users based on their account age (Days)
```

```python
df['cohort0'] = df.apply(lambda row: 1
    if row[
        'accountAge'] < 30
    else '0', axis = 1)
df['cohort1'] = df.apply(lambda row: 1
    if row['accountAge'] >= 30 and row[
        'accountAge'] < 50
    else '0', axis = 1)
df['cohort2'] = df.apply(lambda row: 1
    if row['accountAge'] >= 50 and row[
        'accountAge'] < 365
    else '0', axis = 1)
df['cohort3'] = df.apply(lambda row: 1
    if row['accountAge'] >= 365 and row[
        'accountAge'] < 600
    else '0', axis = 1)
df['cohort4'] = df.apply(lambda row: 1
    if row['accountAge'] >= 600 and row[
        'accountAge'] < 850
    else '0', axis = 1)
df['cohort5'] = df.apply(lambda row: 1
    if row[
        'accountAge'] >= 850
    else 0, axis = 1)

#Print out those columns you want to see
df.loc[0:122, 'cohort0':'cohort5']
```

```python
#For longer logic such as ranking users by their total transaction, 
# and taking their account age cohorts into account
df['RankFrequency'] = df.apply(lambda row: 4
    if ((row['cohort2'] and row['totalCompletedOrders'] == 1) or(row['cohort3'] == 1 and row['totalCompletedOrders'] == 1) or(row['cohort4'] and(row['totalCompletedOrders'] == 1 or row['totalCompletedOrders'] == 2)) or(row['cohort5'] and(row['totalCompletedOrders'] == 1 or row['totalCompletedOrders'] == 2)))
    else(3
        if ((row['cohort1'] == 1 and row['totalCompletedOrders'] == 1) or(row['cohort2'] == 1 and row['totalCompletedOrders'] == 2) or(row['cohort3'] == 1 and row['totalCompletedOrders'] == 2) or(row['cohort4'] == 1 and row['totalCompletedOrders'] == 3) or(row['cohort5'] == 1 and row['totalCompletedOrders'] == 3))
        else(2
            if ((row['cohort1'] == 1 and row['totalCompletedOrders'] == 2) or(row['cohort2'] == 1 and row['totalCompletedOrders'] == 3) or(row['cohort3'] == 1 and row['totalCompletedOrders'] == 3) or(row['cohort4'] == 1 and row['totalCompletedOrders'] == 4) or(row['cohort5'] == 1 and(row['totalCompletedOrders'] == 5 or row['totalCompletedOrders'] == 4)))
            else(1
                if ((row['cohort1'] == 1 and row['totalCompletedOrders'] > 2) or(row['cohort2'] == 1 and row['totalCompletedOrders'] > 3) or(row['cohort3'] == 1 and row['totalCompletedOrders'] >= 4) or(row['cohort4'] == 1 and row['totalCompletedOrders'] >= 5) or(row['cohort5'] == 1 and row['totalCompletedOrders'] >= 6))
                else 0))), axis = 1)

df['RankFrequency']
```

```python
# Lets add in monetary and recency into the mix
df['RankMonetary'] = df.apply(lambda row: 4
    if row['AOV'] < 15
    else(3
        if row['AOV'] >= 15 and row[
            'AOV'] < 25
        else(2
            if row['AOV'] >= 25 and row['AOV'] < 75
            else 1)), axis = 1)

df['RankMonetary']
```

```python
df['RankRecency'] = df.apply(lambda row: 1
    if row['RecencyInTransaction'] <= 31
    else(2
        if row['RecencyInTransaction'] > 31 and row[
            'RecencyInTransaction'] <= 93
        else(3
            if row['RecencyInTransaction'] > 93 and row['RecencyInTransaction'] <= 183
            else 4)), axis = 1)

df['RankRecency']
```

```python
#Now that we have Recency Frequency and Monetary Scoring,"Customer Value Score"
# lets Split the customers into different labelling based on RFM scores.
df['GoldenOpportunity'] = df.apply(lambda row: 1
    if (
        row['indecisiveScoring'] > 1 and row['RankMonetary'] == 1)
    else 0, axis = 1)
df['SilverOpportunity'] = df.apply(lambda row: 1
    if (
        row['indecisiveScoring'] > 1 and row['RankMonetary'] == 2)
    else 0, axis = 1)
df['BronzeOpportunity'] = df.apply(lambda row: 1
    if (row['indecisiveScoring'] > 1 and(
            row['RankMonetary'] == 3) or row['RankMonetary'] == 4)
    else 0, axis = 1)
# The Sure Thing
df['LongTermDecisiveFavourables'] = df.apply(lambda row: 1
    if (row['indecisiveScoring'] < 1 and row[
            'cohort1'] != 1 and row['RankFrequency'] == 1)
    else 0, axis = 1)
df['LongTermDecisiveSplurgers'] = df.apply(lambda row: 1
    if (row['indecisiveScoring'] < 1 and row[
            'cohort1'] != 1 and row['RankFrequency'] <= 2 and row['RankMonetary'] == 1)
    else 0, axis = 1)
df['LongTermDecisiveJoes'] = df.apply(lambda row: 1
    if (row['indecisiveScoring'] < 1 and row[
            'cohort1'] != 1 and row['RankFrequency'] <= 2 and row['RankMonetary'] <= 2)
    else 0, axis = 1)
# The Lost Cause
df['almostUnmarketable'] = df.apply(lambda row: 1
    if (row['indecisiveScoring'] < 1 and row[
            'RankRecency'] == 3 and row['RankFrequency'] == 4)
    else 0, axis = 1)
df['hardToMarketTo'] = df.apply(lambda row: 1
    if (row['indecisiveScoring'] < 1 and row[
            'RankRecency'] == 4 and row['RankFrequency'] == 4)
    else 0, axis = 1)
#These are simple ways to differentiate the various customers
```

```python
# A way to iterate though the customers.
#For this example lets just add customers into list based on their grouping
cohort1 = []
cohort2 = []
cohort3 = []
cohort4 = []
GoldenOpportunity = []
SilverOpportunity = []
BronzeOpportunity = []
LongTermDecisiveFavourables = []
LongTermDecisiveSplurgers = []
LongTermDecisiveJoes = []
almostUnmarketable = []
hardToMarketTo = []
foodAndDrinks = []

for index, row in df.iterrows():
    if row['cohort1'] == 1:
        cohort1.append(row['userId'])
    if row['cohort2'] == 1:
        cohort2.append(row['userId'])
    if row['cohort3'] == 1:
        cohort3.append(row['userId'])
    if row['cohort4'] == 1:
        cohort4.append(row['userId'])
    if row['GoldenOpportunity'] == 1:
        GoldenOpportunity.append(row['userId'])
    if row['SilverOpportunity'] == 1:
        SilverOpportunity.append(row['userId'])
    if row['BronzeOpportunity'] == 1:
        BronzeOpportunity.append(row['userId'])
    if row['LongTermDecisiveFavourables'] == 1:
        LongTermDecisiveFavourables.append(row['userId'])
    if row['LongTermDecisiveSplurgers'] == 1:
        LongTermDecisiveSplurgers.append(row['userId'])
    if row['LongTermDecisiveJoes'] == 1:
        LongTermDecisiveJoes.append(row['userId'])
    if row['almostUnmarketable'] == 1:
        almostUnmarketable.append(row['userId'])
    if row['hardToMarketTo'] == 1:
        hardToMarketTo.append(row['userId'])
    if row['PreferredCategory'] == 'FoodAndDrinks':
        foodAndDrinks.append(row['userId'])


Frequency = {
'cohort1': len(cohort1),
'cohort2': len(cohort2),
'cohort3': len(cohort3),
'cohort4': len(cohort4),
'GoldenOpportunity': len(GoldenOpportunity),
'SilverOpportunity': len(SilverOpportunity),
'BronzeOpportunity': len(BronzeOpportunity),
'LongTermDecisiveFavourables': len(LongTermDecisiveFavourables),
'LongTermDecisiveSplurgers': len(LongTermDecisiveSplurgers),
'LongTermDecisiveJoes': len(LongTermDecisiveJoes),
'almostUnmarketable': len(almostUnmarketable),
'hardToMarketTo': len(hardToMarketTo)
}
#The breakdown of the various cohorts
Frequency
```

```python
#Creating a function to find intersection of multiple list
def intersect( * lists):
    return list(set.intersection( * map(set, lists)))

#users from cohort 2 that prefers foodAndDrinks products
intersect(cohort2,foodAndDrinks)
```

```python
# Next if we got the contacts and email of the customer, we can easily extract them out into a new csv file.
# For instance, we like to send promo email to user that are from cohort 2 and prefers foodAndDrinks voucher.
import random
churnNumber = random.randint(1,101)
def Churn( * lists):
    global churnNumber
    intersectedList = []
    intersectedList = list(set.intersection( * map(set, lists)))
    print("List contains :" + str(len(intersectedList)) + "customers")
    intersectionArray = np.asarray(intersectedList)
    #saves into a new csv file
    df.loc[df['userId'].isin(intersectionArray)].to_csv('extracted' + str(churnNumber) + '.csv')
    churnNumber = churnNumber + 1
    return print("The extracted filename is extracted" + str(churnNumber))

Churn(cohort2,foodAndDrinks)
```
